```python
#读取数据
from pyspark.sql import SparkSession 
spark=SparkSession.builder.appName("dataframe").getOrCreate()
sc = SparkContext.getOrCreate()
data = spark.read.csv(r"1987.csv",header=True,inferSchema=True)
```

```python
from pyspark.sql.types import IntegerType
trans_col = ['DepTime','ArrTime','ActualElapsedTime','AirTime','ArrDelay','DepDelay',
'DepDelay','TaxiIn','TaxiOut','CarrierDelay','WeatherDelay','NASDelay','SecurityDelay','LateAircraftDelay']
for col in trans_col:
    data = data.withColumn(col, data[col].cast(IntegerType()))
data.printSchema() 
```

```python
data = data.withColumn('Distance', data['Distance'].cast('float'))
data.printSchema()
```

```python
data.show(10)
```

```python
#查看每行记录的缺失值
print(data.rdd.map(lambda row:(row['UniqueCarrier'],sum([c==None for c in row]))).collect())
```

```python
#检查每一列中缺失数据的百分比
import pyspark.sql.functions as fn
data.agg(*[(1-(fn.count(c) /fn.count('*'))).alias(c+'_missing') for c in data.columns]).show() 
```

```python
#去掉全是空值的列
#上面可以看出为 1的列有 AirTime、 TaxiIn、TaxiOut,CarrierDelay、WeatherDelay、NASDelay、SecurityDelay、LateAircraftDelay 

drop_list = ['AirTime', 'TaxiOut', 'TaxiIn','CarrierDelay','WeatherDelay','NASDelay','SecurityDelay','LateAircraftDelay' ]
data_drop = data.select([col for col in data.columns if col not in drop_list])
 
 
```

```python
data_drop.agg(*[(1-(fn.count(c) /fn.count('*'))).alias(c+'_missing') for c in data_drop.columns]).show() 
```

统计分析

```python
#交叉表，看月份和延迟起飞时间的交叉报表
data_drop5.crosstab('Month', 'DepDelay').show()
```

```python
#看不同公司的延迟到达平均时间
data_drop5.groupby('UniqueCarrier').agg({'ArrDelay': 'mean'}).show()
```

```python
data_drop5.groupby('UniqueCarrier').count().show()
```

```python
#航班的平均延迟时间

averageDelays = data_drop5.groupby('FlightNum').agg({'ArrDelay': 'mean'} ).show()
 
```

```python
from pyspark.sql import functions as fn
averageDelays = data_drop5.groupBy(data_drop5.FlightNum).agg(fn.avg(data_drop5.ArrDelay), fn.avg(data_drop5.DepDelay))
```

```python
averageDelays.show(30)
```

最后的5个特征都是一种延误分类，那么可以思考有哪些因素导致延误，需要增加哪些可测的数据因素，从而用特征来预测飞机是否延误，延误时间等。
